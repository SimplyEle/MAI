```python
from typing import Tuple, Mapping

import numpy as np
from matplotlib import pyplot as plt
import sklearn
import pygmo as pg
from PIL import Image
import requests
```

```python
def check_distance(x0, x_calc):
  return np.sqrt((x_calc[0]**2-x0[0])+(x_calc[1]**2-x0[1]))
```

### Тестовые функции

##### Функция трехгорбого верблюда
$$ {\displaystyle f(x,y)=2x^{2}-1.05x^{4}+{\frac {x^{6}}{6}}+xy+y^{2}} $$
Глобальный минимум
$$ {\displaystyle f(0,0)=0} $$
Метод поиска
$$ {\displaystyle -5\leq x,y\leq 5} $$

```python
class TestFuncCamel:
  def __init__(self, x_min, x_max, y_min, y_max):
    self.x_min = x_min
    self.x_max = x_max
    self.y_min = y_min
    self.y_max = y_max

  def fitness(self, x) -> np.float64:   
    return [2 * x[0]**2 - 1.05 * x[0]**4 + 1/6 * x[0]**6 + x[0] * x[1] + x[1]**2]

  def get_bounds(self):
    return ([self.x_min, self.y_min], [self.x_max,self.y_max])
```

```python
camel = pg.problem(TestFuncCamel(-5,5,-5,5))
print(camel)
```

##### Функция Изома
$$ {\displaystyle f(x,y)=-\cos \left(x\right)\cos \left(y\right)\exp \left(-\left(\left(x-\pi \right)^{2}+\left(y-\pi \right)^{2}\right)\right)} $$
Глобальный минимум
$$ {\displaystyle f(\pi ,\pi )=-1} $$
Метод поиска
$$ {\displaystyle -100\leq x,y\leq 100} $$

```python
class TestFuncIzom:
  def __init__(self, x_min, x_max, y_min, y_max):
    self.x_min = x_min
    self.x_max = x_max
    self.y_min = y_min
    self.y_max = y_max
    
  def fitness(self, x) -> np.float64:   
      return [-np.cos(x[0]) * np.cos(x[1]) * np.exp(-((x[0] - np.pi)**2 + (x[1] - np.pi)**2))]

  def get_bounds(self):
    return ([self.x_min, self.y_min], [self.x_max,self.y_max])
```

```python
izom = pg.problem(TestFuncIzom(-100,100,-100,100))
print(izom)
```

##### Функция Била
$$ {\displaystyle f(x,y)=\left(1.5-x+xy\right)^{2}+\left(2.25-x+xy^{2}\right)^{2}} $$
Глобальный минимум:
$$ {\displaystyle f(3,0.5)=0} $$
Метод поиска:
$$ {\displaystyle -4.5\leq x,y\leq 4.5} $$

```python
class TestFuncBil:
  def __init__(self, x_min, x_max, y_min, y_max):
    self.x_min = x_min
    self.x_max = x_max
    self.y_min = y_min
    self.y_max = y_max
    
  def fitness(self, x) -> np.longdouble:
      return [(1.5 - x[0] + x[0]*x[1])**2 + (2.25 - x[0] + x[0]*x[1]**2)**2]
  
  def get_bounds(self):
    return ([self.x_min, self.y_min], [self.x_max,self.y_max])
```

```python
bil = pg.problem(TestFuncBil(-4.5,4.5,-4.5,4.5))
print(bil)
```

### Алгоритм Particle Swarm Optimization (PSO) - Метод роя частиц
#### Generational (also problems deriving from base_stochastic)

```python
num_of_evols = 100

def evol_f_pso(func):
  alg = pg.algorithm(pg.pso_gen())
  popul = pg.population(func, size = 1000)

  

  fit_list = []
  ind_list = []

  for _ in range(num_of_evols):
    popul = alg.evolve(popul)
    ind_list.append(popul.get_x()[popul.best_idx()])
    fit_list.append(popul.get_f()[popul.best_idx()])
  return fit_list, ind_list
```

```python
def do_plot_champ(num_of_evols, fit_list):
  fig, sp = plt.subplots(figsize=(10,10))
  sp.plot(np.arange(0, num_of_evols), fit_list)

  champ = np.argmin(np.array(fit_list))
  sp.scatter(champ, np.min(fit_list), marker='o', color='g')

  sp.set_xlim((0,num_of_evols))
  sp.grid('major')
  sp.set_title('Champion of every evol')
  sp.set_xlabel("Num of evol")
  sp.set_ylabel("F")

  plt.show()
```

```python
champions = [[0,0,0],[0,0,0]]
```

```python
fit_list_camel, ind_list_camel = evol_f_pso(camel)
```

```python
x_calc_camel = ind_list_camel[np.argmin(np.array(fit_list_camel))]

champions[0][0] = check_distance([0,0],x_calc_camel)
```

```python
fit_list_izom, ind_list_izom = evol_f_pso(izom)
```

```python
x_calc_izom = ind_list_izom[np.argmin(np.array(fit_list_izom))]

champions[0][1] = check_distance([np.pi,np.pi],x_calc_izom)
```

```python
fit_list_bil, ind_list_bil = evol_f_pso(bil)
```

```python
x_calc_bil = ind_list_bil[np.argmin(np.array(fit_list_bil))]

champions[0][2] = check_distance([3,0.5],x_calc_bil)
```

```python
class TestImage:
  
  def __init__(self, path_of_image):
    self.image = np.asarray(Image.open(path_of_image))
    self.x_min = 0 
    self.y_min = 0
    self.x_max = self.image.shape[1]
    self.y_max = self.image.shape[0]
    

  def fitness(self, x):
    return [sum(self.image[int(x[0])][int(x[0])])]

  def get_bounds(self):
    return ([self.x_min, self.y_min], [self.x_max, self.y_max])
```
